```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,50)
y = 2*x + 1
#plt.figure():定义一个图形窗口
plt.figure()
#plt.plot(x,y):画简单的直线图
plt.plot(x,y)
#plt.show():一系列操作后，打印出图片
plt.show()
```

```python
x = np.linspace(-3,3,50)
y1 = 2*x + 1
y2 = x**2
plt.figure()
#连续画两条线，可以在一个plt.figure()下作图，最后作图结束用plt.show()打印
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')
plt.show()
```

```python
x = np.linspace(-3,3,50)
y1 = 2*x + 1
y2 = x**2
plt.figure()
#连续画两条线，可以在一个plt.figure()下作图，最后作图结束用plt.show()打印
plt.plot(x,y2)
plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--')
#设置x和y的坐标区间
plt.xlim((-1,2))
plt.ylim((-2,3))
#改变x和y坐标轴的刻度，甚至可以用字符替代坐标轴刻度
plt.xticks(np.linspace(-1,2,5))
plt.yticks([-2,-1.8,-1,1.22,3],[r'$really\ bad$','$bad$','$normal$','$good$',r'$really\ good$'])
#设置x和y坐标轴的标题
plt.xlabel(('i am x'))
plt.ylabel(('i am y'))
#使用plt.gca()获取当前坐标轴信息
ax = plt.gca()
#使用ax.spines设置边框，这里是将边框去除
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#使用ax.xaxis.set_ticks_position来设置x坐标轴客户名字或名称的位子(可选项：top,bottom,both,default,none)
ax.xaxis.set_ticks_position('bottom')
#使用ax.spines设置边框：x轴；.set_position设置边框位置：y=0的位置（可选项：outward,axes,data）
ax.spines['bottom'].set_position(('data',0))
#使用ax.spines设置边框：y轴；.set_position设置边框位置：x=0的位置（可选项：outward,axes,data）
ax.spines['left'].set_position(('data',0))
plt.show()
```

```python
x = np.linspace(-3,3,50)
y1 = 2*x + 1
y2 = x**2
plt.figure()
#连续画两条线，可以在一个plt.figure()下作图，最后作图结束用plt.show()打印
l1, = plt.plot(x,y2,label='linear line')
l2, = plt.plot(x,y1,color='red',linewidth=1.0,linestyle='--',label='square line')
#设置x和y的坐标区间
plt.xlim((-1,2))
plt.ylim((-2,3))
#改变x和y坐标轴的刻度，甚至可以用字符替代坐标轴刻度
plt.xticks(np.linspace(-1,2,5))
plt.yticks([-2,-1.8,-1,1.22,3],[r'$really\ bad$','$bad$','$normal$','$good$',r'$really\ good$'])
#使用legend添加图例，参数'upper right'表示图例将添加在图中的右上角
plt.legend(loc='upper right')
plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure(num=1, figsize=(8, 5),)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
plt.plot(x, y)

#在图形中标记出(1,3)点的位置信息
x0 = 1
y0 = 2*x0 + 1
#这里其实画了一条新的线段
plt.plot([x0,x0],[0,y0],'k--',linewidth=2.5)
#这里画了一个新的点，plt.scatter用来用散点图的
plt.scatter([x0],[y0],s=50,color='b')

#plt.annotate添加注释,xycoords='data'是说基于数据的值来选位置，xytext=(+30,-30),textcoords='offset points'
#对于标注位置的描述和xy偏差值，arrowprops是对图中箭头类型的一些设置
plt.annotate(r'$2x+1=%s$' % y0,xy=(x0,y0),xycoords='data',xytext=(+30,-30),
            textcoords='offset points',fontsize=16,
            arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))
#plt.text添加注释,-3.7,3,是选取text的位置，空格需要用到转字符\，fontdict设置文本字体
plt.text(-3.7,3,r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',fontdict={'size':16,'color':'r'})
plt.show()
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)
#plt.scatter()用来画散点图
plt.scatter(X,Y,s=75,c=T,alpha=.5)
plt.xlim(-1.5,1.5)
plt.xticks(())#ignore xticks
plt.ylim(-1.5,1.5)
plt.yticks(())
plt.show()
```

```python
n = 12
X = np.arange(n)
Y1 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)
#plt.bar用来画柱状图
plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')

plt.xlim(-.5,n)
plt.xticks(())
plt.ylim(-1.25,1.25)
plt.yticks(())
#在图中加上数值.ha='center'横向对齐,va='bottom'纵向底部对齐
for x,y in zip(X,Y1):
    plt.text(x,y+0.05,'%.2f' % y,ha='center',va='bottom')
for x,y in zip(X,Y2):
    plt.text(x,-y-0.05,'%.2f' % y,ha='center',va='top')

plt.show()
```

```python
def f(x,y):
    return (1 - x/2 + x**5 + y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
#plt.contourf用来进行颜色填充
plt.contourf(X,Y,f(X,Y),8,alpha=.75,cmap=plt.cm.hot)
#plt.contour函数划线
C = plt.contour(X,Y,f(X,Y),8,colors='black',linewidth=.5)
plt.clabel(C,inline=True,fontsize=10)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y) #x-y平面的网格
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap = plt.get_cmap('rainbow'))
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap=plt.get_cmap('rainbow'))
plt.show()
```
